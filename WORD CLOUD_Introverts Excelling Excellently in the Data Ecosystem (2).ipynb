```python
# Import the Required Libraries:

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
```

```python
# Sample text and visualization using Word Cloud

text = "/Volumes/Expansion/Documents/My Digital Portfolio/Word Cloud/Thriving in the Field of Data Science as an Introvert.txt"

python_mask = np.array(PIL.Image.open("/Volumes/Expansion/Documents/My Digital Portfolio/AI_Prostock-studio_Alamy.jpeg"))

colormap = ImageColorGenerator(python_mask)

wc = WordCloud(stopwords=STOPWORDS, 
               mask=python_mask, 
               background_color="white",
               contour_color="black", 
               contour_width=3, 
               min_font_size=3, 
               max_words=400).generate(text)

wc.recolor(color_func=colormap)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

## INTERACTIVE WORD CLOUD

Here, I created an inteactive word clound that shows the data skill in deman per year.
If you select a year, it interacts and display the most sought after data skill for that year.

### Steps
1. Name the variables (Please note that the data was only assumed)
2. Create a dropdown and the required options.
3. Create a function that will display the word cloud.
4. Create the widgets that will interact with the function.

```python
#Data Skill in Demand in Years 2020, 2021 and 2022

# Naming the variables.

Year_2018 = {'Excel':24, 'R':19, 'Apache':15, 'Java':1, 'SQL':3, 'Python':1}
Year_2020 = {'Excel':17, 'R':17, 'Apache':8, 'Java':9, 'SQL':2, 'Python':3}
Year_2022 = {'Excel':10, 'R':2, 'Apache':8, 'Java':6, 'SQL':12, 'Python':13}

# Creatimng a dropdown.

drop_down = ipywidgets.Dropdown(options=[('2018', Year_2018), ('2020', Year_2020), ('2022', Year_2022)],
                               value=Year_2018,
                               description='Year')
#Creating the function

def update_wc(choice):
    '''Create a function that will display the word cloud.'''
    wc = WordCloud(colormap='tab10').generate_from_frequencies(choice)
    plt.figure(figsize=(8,8))
    plt.imshow(wc)
    plt.axis("off")

    
ipywidgets.interact(update_wc, choice=drop_down);
```
